```python
delivery_df["current_score"] = delivery_df.groupby("match_id")["total_runs_y"].cumsum()
delivery_df["current_score"]
```

```python
delivery_df["runs_left"]=delivery_df["total_runs_x"]-delivery_df["current_score"]
delivery_df
```

```python
delivery_df["balls_left"]=120- (delivery_df["over"]*6 +delivery_df["ball"])
delivery_df
```

```python
# Step 1: Replace NaN with "0"
delivery_df["player_dismissed"] = delivery_df["player_dismissed"].fillna(0)

# Step 2: Map values: 0 => 0 (no wicket), name => 1 (wicket)
delivery_df["player_dismissed"] = delivery_df["player_dismissed"].apply(lambda x: 0 if x == 0 else 1)

# Step 3: Now it is numeric, so we can cumsum
delivery_df["wickets_fallen"] = delivery_df.groupby("match_id")["player_dismissed"].cumsum()

# Step 4: Remaining wickets
delivery_df["wickets_left"] = 10 - delivery_df["wickets_fallen"]

# View the result
delivery_df.head()
```

```python
delivery_df["crr"]=delivery_df["current_score"]/(120-delivery_df["balls_left"])
delivery_df["rrr"]=(delivery_df["runs_left"]*6)/delivery_df["balls_left"]
delivery_df
```

```python
def result(row):
    return 1 if row["batting_team"]==row["winner"] else 0
    
delivery_df["result"]=delivery_df.apply(result,axis=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ("trf", OneHotEncoder(sparse_output=False, drop="first"), ["batting_team", "bowling_team", "city"])
], remainder="passthrough")
```

```python
import numpy as np

def match_progression(x_df, match_id, pipe):
    # Filter match
    match = x_df[x_df['match_id'] == match_id]
    match = match[match['ball'] == 6]  # Only take end of over deliveries
    
    # Select important columns
    temp_df = match[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wickets_left', 'crr', 'rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    
    # Predict probabilities
    result = pipe.predict_proba(temp_df)
    
    temp_df['lose'] = np.round(result.T[0]*100, 1)  # Losing probability
    temp_df['win'] = np.round(result.T[1]*100, 1)   # Winning probability
    temp_df['end_of_over'] = range(1, temp_df.shape[0] + 1)

    # Handling target and run calculation
    target = temp_df['runs_left'].iloc[0] + (temp_df['crr'].iloc[0] * (temp_df['balls_left'].iloc[0] / 6))
    target = int(target)  # Convert to int if needed

    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0, target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)

    # Handling wickets
    wickets = list(temp_df['wickets_left'].values)
    new_wickets = [10] + wickets[:-1]  # shift properly

    w = np.array(wickets)
    nw = np.array(new_wickets)

    temp_df['wickets_in_over'] = (nw - w)


    print("Target -", target)
    
    temp_df = temp_df[['end_of_over', 'runs_after_over', 'wickets_in_over', 'lose', 'win']]
    
    return temp_df, target
```
